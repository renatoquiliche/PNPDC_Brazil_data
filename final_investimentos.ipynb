```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
df = pd.read_excel('investimentos_ibge_v2.xlsx')
df.columns
```

```python
# Que tem municipio
municipio = df.loc[df['COD IBGE'].dropna().index]
municipio = municipio.groupby(['Ano', 'COD IBGE', 'UF', 'Tipo', 'Eixo de Atuação'])['Valor total pago/transFérido'].agg('sum')
municipio = municipio.reset_index()
municipio
```

```python
# Que tem so estado e nao tem municipio
estado = df.loc[df['COD IBGE'].isna() & (df['UF'].isna() == False)]
#print((estado.loc[estado.UF == 'RJ']) AND (estado.Ano == 2014))
estado = estado.groupby(['Ano', 'UF', 'Tipo', 'Eixo de Atuação'])['Valor total pago/transFérido'].agg('sum')
estado = estado.reset_index()
estado.loc[estado['UF'] == 'RO']
```

```python
aux_data = pd.read_excel('C:/Users/xx-re/Downloads/Separação Investimentos.xlsx', sheet_name='Planilha1')
aux_data = aux_data[['Cod IBGE', 'UF', 'Município', 'prop']]
aux_data.loc[:,'COD UF'] = aux_data['Cod IBGE'].astype('str').str.slice(0,2)
aux_data
```

```python
aux_data.loc[aux_data['Cod IBGE'] == 1100296]
```

```python
#Primeiro calculamos o total estadual
estado = estado.groupby(['Ano', 'UF', 'Tipo', 'Eixo de Atuação'])['Valor total pago/transFérido'].agg('sum')
#pd.merge(estado, aux_data, left_on='UF', right_on='UF')
estado = estado.reset_index()
```

```python
estado = pd.merge(estado, aux_data, left_on='UF', right_on='UF')
estado['valor_investimento'] = estado['Valor total pago/transFérido']*estado['prop']
estado.loc[estado['Cod IBGE']==1100296]
```

```python
estado = estado.groupby(['Ano', 'Cod IBGE', 'Tipo', 'Eixo de Atuação'])['valor_investimento'].agg('sum').reset_index()
estado
```

```python
municipio = municipio.groupby(['Ano', 'COD IBGE', 'Tipo', 'Eixo de Atuação'])['Valor total pago/transFérido'].agg('sum').reset_index()

municipio.rename(columns={'COD IBGE': 'Cod IBGE', 'Valor total pago/transFérido':'valor_investimento'}, inplace=True)
municipio.shape
```

```python
final = pd.concat([estado, municipio])
#final.loc[final['Cod IBGE']==3303906]
final = final.groupby(['Ano', 'Cod IBGE', 'Tipo', 'Eixo de Atuação'])['valor_investimento'].agg('sum').reset_index()
```

```python
final = final.pivot_table(index=['Ano', 'Cod IBGE', 'Eixo de Atuação'], columns='Tipo', values='valor_investimento').reset_index().fillna(0)
final.loc[final['Cod IBGE']==3303906]
```

```python
final.to_excel('final_data_investimentos.xlsx', index=False)
```
