```python
import pandas as pd
```

```python
mun = pd.read_excel("../Data/população_investimentos_v2.xlsx", sheet_name='MUNICIPIO')
```

```python
estado = pd.read_excel("../Data/população_investimentos_v2.xlsx", sheet_name='Planilha2')
```

```python
prop1 = pd.merge(mun,estado, on='COD. UF')
prop1['prop1'] = prop1['POP. TOTAL_x']/prop1['POP. TOTAL_y']
```

```python
mun['POP BR'] = estado['POP. TOTAL'][0]
```

```python
prop2 = mun['POP. TOTAL'] / mun['POP BR']
```

```python
prop1
```

```python
prop_final = prop1[['UF_x', 'Cod IBGE', 'NOME DO MUNICÍPIO', 'prop1']]
prop_final.loc[:,['prop2']] = prop2
```

```python
prop_final.rename(columns={'NOME DO MUNICÍPIO':'Município', 'UF_x':'UF'}, inplace=True)
```

```python
prop_final
```

```python
prop_final.to_excel('../Data/proportions.xlsx', index=False)

# Use prop1 when there is ESTADO or MUNICIPIO
# Use prop2 when there is neither ESTADO nor MUNICIPIO
```
