```python
aux_data = pd.read_excel('C:/Users/xx-re/Downloads/Separação Investimentos.xlsx', sheet_name='Planilha1')
aux_data = aux_data[['Cod IBGE', 'UF', 'Município', 'prop']]
aux_data.loc[:,'COD UF'] = aux_data['Cod IBGE'].astype('str').str.slice(0,2)
aux_data
```

```python
aux_data = pd.read_excel('../Data/proportions.xlsx')
aux_data = aux_data[['Cod IBGE', 'UF', 'Município', 'prop1', 'prop2']]
aux_data.loc[:,'COD UF'] = aux_data['Cod IBGE'].astype('str').str.slice(0,2)
aux_data
```

```python
#Primeiro calculamos o total estadual
estado = estado.groupby(['Ano', 'UF', 'Tipo', 'Eixo de Atuação_rec'])['Valor total pago/transFérido'].agg('sum')
#pd.merge(estado, aux_data, left_on='UF', right_on='UF')
estado = estado.reset_index()
```

```python
# unimos con aux_data para llenar los municipios
estado = pd.merge(estado, aux_data, left_on='UF', right_on='UF')
estado['valor_investimento'] = estado['Valor total pago/transFérido']*estado['prop1']
estado.loc[estado['Cod IBGE']==1100296]
```

```python
estado = estado.groupby(['Ano', 'Cod IBGE', 'Tipo', 'Eixo de Atuação_rec'])['valor_investimento'].agg('sum').reset_index()
estado
```

# Unimos los grupos estado y municipio, aqui unimos tambien con union

```python
final = pd.concat([estado, municipio, brazil])
#final.loc[final['Cod IBGE']==3303906]
final = final.groupby(['Ano', 'Cod IBGE', 'Tipo', 'Eixo de Atuação_rec'])['valor_investimento'].agg('sum').reset_index()
```
